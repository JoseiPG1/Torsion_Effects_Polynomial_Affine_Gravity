```python
import numpy as np
import matplotlib.pyplot as plt

# Function for the scale factor
def scale_factor(t, alpha, beta, gamma):
    return np.sqrt(-((alpha**2 * beta * (2 * alpha * beta - 3 * gamma) * (t * (alpha * beta + gamma))**((2 * alpha * beta)/(alpha * beta + gamma))) / (2 * t**2 * (alpha * beta + gamma)**2)))

# Time range for the plot
t_values = np.linspace(0.1, 10, 100)  # Adjust the range as needed

# Different sets of alpha, beta, and gamma values
alpha_values = [-1, -2, -3]
beta_values = [-3, -6, -9]
gamma_values = [-alpha * beta / 3 for alpha, beta in zip(alpha_values, beta_values)]

# Plotting for different sets of alpha, beta, and gamma
plt.figure(figsize=(8, 6))

for alpha, beta, gamma in zip(alpha_values, beta_values, gamma_values):
    a_values = scale_factor(t_values, alpha, beta, gamma)
    plt.plot(t_values, a_values, label=f'α={alpha}, β={beta}, γ={gamma}')

# Adding labels and title
plt.xlabel("Time")
plt.ylabel("Affine Scale Factor $a_f(t)$")
plt.title("Affine Radiation for different parameters")
plt.legend()
plt.grid(True)
plt.savefig('Affine_Radiation.pdf')
plt.show()

```
